## Vorbereitender Schritt - Identifikation von Typos be der Erfassung

```python
# Finde in der Spalte "Wohnort", Gemeinden, die sich nur in einem Buchstaben unterscheiden

gemeinden = df_cars.groupby("Kreis_Stadt").Wohnort.unique()

for key, locations in gemeinden.items():
    for i in locations:
        for j in locations:
            if i<j:
                s = SequenceMatcher(None, i, j)
                if s.ratio() > 0.9:
                    print(f"{i} ({df_cars[df_cars.Wohnort == i].Kreis_Stadt.iloc[0]}) , {j} ({df_cars[df_cars.Wohnort == j].Kreis_Stadt.iloc[0]})")
```
